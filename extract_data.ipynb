```python
from yaml import safe_load
import os
from tqdm import tqdm
```

```python
filename=[]
for file in os.listdir('t20s'):
    filename.append(os.path.join('t20s',file))
```

```python
filename[0:5]
```

```python
data=pd.DataFrame()
c=1
for file in tqdm(filename):
    with open(file,'r') as f:
        da=pd.json_normalize(safe_load(f))
        da['match_id']=c
        data=data.append(da)
        c+=1
```
